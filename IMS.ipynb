```python
import json

fd = open("records.json","r");
txt = fd.read()
fd.close()

record = json.loads(txt)
print("\t\t\tMENU\n");
print("UID"+"\t\t"+"Product"+"\t\t"+"Quantity"+"\t"+"Price\n");
for i in record:
    print(i,"\t\t", record[i]['name'],"\t\t",record[i]['quantity'],"\t\t",record[i]['price']);

    
flag = 1;
sum1 = 0;
while(flag):
    print("\n\nPlease select Products UID no. to buy And -1 for exit\n");
    x = input("S.NO- ");
    if(x == "-1"):
        print("\n\n\t\tThank you for shopping\n");
        break;
        
        
    y = int(input("Enter Quantity - "));
    
    
    
    if(y > record[x]['quantity']):
        print("Sorry ! Available Quantity is only ",record[x]['quantity']);
        continue;
    
    print("\nUID"+"\t\t"+"Product"+"\t\t"+"Rs.* Price"+"\t  "+"Total\n");
    print(x,"\t\t", record[x]['name'],"\t\t",record[x]['price']," X ",y,"       ", record[x]['price']*y);
    
    record[x]['quantity'] = record[x]['quantity'] - y;
    sum1 = sum1 + record[x]['price']*y;
    
    
    record1 = json.dumps(record);
    fd = open('records.json','w')
    fd.write(record1)
    fd.close();
    
    print("\n\t\t\tUpdated MENU\n")
    for i in record:
        print(i,"\t\t", record[i]['name'],"\t\t",record[i]['quantity'],"\t\t",record[i]['price']);

print("Total bill - ",sum1);



```
